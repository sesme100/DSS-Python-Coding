## git 안에 파일디렉토리 My first repo 생성함 


$ git status
## 상태확인

$ touch 파일이름.py
## 내 로컬 서버 (내문서, dev파일에 파일을 하나 생성) 

$ vim 파일이름.py 

## vim 프로그램에서 내 파일안에 내용을 생성
$ python 파일이름.py
## 파이썬에서 파일이 잘 작동하는지 확인


$ git init 
$ git add 파일이름 

## git 내 서버에 파일을 add 
$ git commit -m "some commit파일이름"
## commit 하기
$ git remote add origin https://github.com/username/repo.git 
## origin 은 뒷 웹주소의 별명 
$ git push origin master
##푸쉬 하기 

$ git pull origin master
## 만약 업데이트가 git directory 에서만 되었다면 git pull 해서 git 에서도 업데이트 된 사항들을 불러와서 확인해야 그다음 명령을 먹음
