```python
## get/set 메소드 지정 

class Cal(object):
    def __init__(self,v1,v2):
        if isinstance(v1, int) and isinstance(v2, int):
            self.v1 = v1## v1은 인스턴스의 벨류값
            self.v2 = v2
        #print(v1,v2)
        ## 여기서 self는 들어온 인스턴스 가르킴

    def add(self):
        return self.v1+self.v2

    def subtract(self): ## 여기서 self는 들어온 인스턴스 가르킴
        return self.v1-self.v2

########################################  get/setValue 설정하기(class내에서 인스턴스 변수 다시 설정하기)
    def show(self): ## 지금 value들을 보여주기  
        print(self.v1, self.v2)

    def getValue(self): ## 지금 설정되어진 value들을 불러오기 
        return self.v1, self.v2

    def setValue(self, newValue1, newValue2):## 새롭게 value들을 설정하기
        if isinstance(newValue1, int) and isinstance(newValue2,int):
            self.v1 = newValue1
            self.v2 = newValue2

F1 = Cal(10,10)
print(F1.getValue())

F1.setValue('one',30) 
## 'one'이 string이기 때문에 isinstance에서 걸려서  False 가 발생해서 무시됨 그래서 기존 10,10의 값이 나옴
print(F1.getValue())
F1.show()

################################ 하기에는 외부에서 value 설정방법(추천하지 않음)
#C1 = Cal(10,10)
#print(C1.add())
#print(C1.subtract())

#C2 = Cal(20,30)
#print(C2.add())
#print(C2.subtract())

#C3 = Cal(50,70)
# print(C3.v1, C3.v2) 
# ## C3 의 Value 변경함 (파이써는 클래스 외부 변수를 가져와서 셋팅가능)
#C3.v1 = 80
#C3.v2 = 90
#print(C3.v1, C3.v2) ## value 변경 외부의 벨류변경을 막을수 없음 

'''객체 = 부품
캡슐화 ( encapsulation) : 변수가 외부의 영향을 덜받게끔해주어서 전체적으로 안정적으로 만듬 

데이터 VS. 로직
데이터가 더 훨씬 중요한데 캡슐화를 하면 데이터를 조금더 안전하게 저장할수 있음
--------------------------------------------------------------------
객체지향
get/set method : 파이썬의 경우 외부 value가 인스턴스 변수에서 변경가능하기 때문에 get/ set 메소드가 권장됨
즉, 외부에서 인스턴스 변수를 읽고 쓰기 접근 하기가 가능 

메소드 바깥쪽에서 접근할 수 없는 변수가 된다는 말은, 해당 변수가 속해 있는 객체 내의 다른 모든 메소드에서 조차도 변수에 접근할 수가 없다는 얘기가 됩니다. 

instance바깥쪽에서 접근할 수 없는 변수가 된다 가 맞는 말입니다. (또한 객체지향 프로그래밍 강좌 초반 부분에서 인스턴스와 객체는 같은(정확히 같진 않겠지만) 것이라고 말씀하셨습니다.)
'''
```

```python
## 상속만들기 inheritance 

class Cal(object):
    def __init__(self,v1,v2):
        self.v1 = v1## v1은 인스턴스의 벨류값
        self.v2 = v2
        #print(v1,v2)
        ## 여기서 self는 들어온 인스턴스 가르킴

    def add(self):
        return self.v1+self.v2

    def subtract(self): ## 여기서 self는 들어온 인스턴스 가르킴
        return self.v1-self.v2

##################################### 상속만들기 inheritance 
class Cal3(Cal): ## Cal을 상속받아 Cal3이라는 클래스 만들기 
    
    def divide(self):## 새로운 인스턴스 매소드 추가함
        return self.v1/self.v2 

    def multi(self):## 새로운 인스턴스 매소드 추가함
        return self.v1*self.v2



C1 = Cal(10,10)
print(C1.add())
print(C1.subtract())
## print(C1.divide())## Error: C1은 Cal 만 사용하는 인스턴스 이므로 에러남

C2 = Cal3(20,40) 
## Cal을 상속받은 Cal3는 추가한 메소드 포함하여 모든 메소드 사용가능
## 만약 Cal 의 내용이 바뀌면 상속받은 Cal3에 원조 메소드 바뀐내용도 적용된다 
print(C2.add())
print(C2.subtract())
print(C2.divide()) ## Cal3 에서는 Cal의 메소드와 추가 메소드 사용가능 
print(C2.multi())

'''상속(Inheritance) 

기존의 클래스에서 기존의 메소드에  새로운 메소드를 추가하여 추가 결합한 것을 상속이라고 한다 

'''
```

```python
## ClassMethod 

class Cal(object):
    _history=[] ## 클래스 변수 설정(_ 때문에 클래스 내부에서만 사용가능)

    def __init__(self,v1,v2):
        if isinstance(v1, int) and isinstance(v2, int):
            self.v1 = v1## v1은 인스턴스의 벨류값
            self.v2 = v2
        #print(v1,v2)
        ## 여기서 self는 들어온 인스턴스 가르킴

    def add(self):
        result = self.v1+self.v2
        Cal._history.append("ADD: "+str(result)) ## 클래스의 히스토리 메소드에 add결과값 넣기 
        return result

    def subtract(self): ## 여기서 self는 들어온 인스턴스 가르킴
        result = self.v1-self.v2
        Cal._history.append("Subtract: "+str(result))
        return result

    @classmethod ## 클라스 매소드 설정
    def history(cls): ## 꼭 클래스 매개변수 꼭 넣기
        for item in Cal._history: ##_는 클래스 내부에서만 사용할 메소드라는 의미
            print(item) ## Cal._history에 append된 내용들이 다나옴
        ## 이거 하나만 실행 프린트 하면 모든 append된 내용들이 다 나오는것임 

class Cal3(Cal):
    def multi(self):
        result = self.v1*self.v2
        Cal._history.append("Multiply: "+str(result))
        return result

    def divide(self):
        result = self.v1/self.v2
        Cal._history.append("Divide: "+str(result))
        return result
    

F1 = Cal(10,10)
F1.add()
F1.subtract()

F2 = Cal(20,20)
F2.add()
F2.subtract()

F3 = Cal3(30,60)
F3.add()
F3.subtract()
F3.multi()
F3.divide()

Cal.history() ## 클래스 메소드 불러오기 ## 클라스 호출 되어 실행된 history 나타냄
## 반드시 add와 subtract 메소드가 사용이 한번씩 되어야 하므로 인스턴스 생성하여 써보기 
## 그렇게 해야지만 클라스 history 메소드에 append가 되어 결과값 출력 
```

```python
## Over Ride: 재정의 

## 상속 : 부모객체의 method들을 물려받아서 자식 객체가 부모객체의 method들을 사용 가능 + 자식 객체 method를 추가하여 사용
## Over Ride(재정의): 부모객체의 Method중에서 맘에 들지 않는 method를 재정의 하여 사용 

class Cal(object):
    _history=[] ## 클래스 변수 설정(_ 때문에 클래스 내부에서만 사용가능)

    def __init__(self,v1,v2):
        if isinstance(v1, int) and isinstance(v2, int):
            self.v1 = v1## v1은 인스턴스의 벨류값
            self.v2 = v2
        #print(v1,v2)
        ## 여기서 self는 들어온 인스턴스 가르킴

    def add(self):
        result = self.v1+self.v2
        Cal._history.append("ADD: "+str(result)) ## 클래스의 히스토리 메소드에 add결과값 넣기 
        return result

    def subtract(self): ## 여기서 self는 들어온 인스턴스 가르킴
        result = self.v1-self.v2
        Cal._history.append("Subtract: "+str(result))
        return result

    @classmethod ## 클라스 매소드 설정
    def history(cls): ## 꼭 클래스 매개변수 꼭 넣기
        for item in Cal._history: ##_는 클래스 내부에서만 사용할 메소드라는 의미
            print(item) ## Cal._history에 append된 내용들이 다나옴
        ## 이거 하나만 실행 프린트 하면 모든 append된 내용들이 다 나오는것임 
    
    def info(self): ## 부모클래스 메소드 
        return "Cal => v1: %d, v2: %d" %(self.v1, self.v2)


class Cal3(Cal):
    def multi(self):
        result = self.v1*self.v2
        Cal._history.append("Multiply: "+str(result))
        return result

    def info(self):
        return "I am child class's Method => "+ super().info() ## 자식 클래스 + 부모클래스의 info method실행 


    def divide(self):
        result = self.v1/self.v2
        Cal._history.append("Divide: "+str(result))
        return result
    

F0 = Cal(30,60)
print(F0.info()) ## Cal => v1: 30, v2: 60

F1 = Cal3(10,10)
print(F1.info()) ## I am child class's Method => Cal => v1: 10, v2: 10


```

```python
## 다중상속(Multiple Inheritance)
## 여러개의 부모 클래스의 메소드를 자식클래스가 가짐
## 다중상속의 단점: 문제가 많이 생김 
## Ruby 언어의 믹스인과 비슷 

class C1():
    def __init__(self,v1,v2):
        self.v1 = v1## v1은 인스턴스의 벨류값
        self.v2 = v2 

    def add(self):
        return self.v1+self.v2

    def subtract(self): ## 여기서 self는 들어온 인스턴스 가르킴
        return self.v1-self.v2
         
class C2():
    def __init__(self,v1,v2):
        self.v1 = v1## v1은 인스턴스의 벨류값
        self.v2 = v2

    def divide(self):## 새로운 인스턴스 매소드 추가함
        return self.v1/self.v2 

    def multi(self):## 새로운 인스턴스 매소드 추가함
        return self.v1*self.v2


class C3(C1, C2): ## 부모클래스 C1과 C2 모두 상속받음 ## 순서에따라 C1 이 우선됨
    pass ## C3에 아무런 내용이 없는경우 


child = C3(25,35)
print(child.add())
print(child.divide())
print(child.subtract())
print(child.multi())
```

```python
## Crawling Library 사용하여 크롤러 만들기
''' 
1. 인터넷에 python3 http downlad library 검색
2. 윈도우 cmd 키고 request라는 library를 다운받고 싶어하는 디렉토리 가기
3. 거기서 pip install requests 다운받기 
4. import requests 하기 
5. 다운받고 싶어하는 웹 홈페이지 주소 넣기
6. html text 파일로 인쇄 
7. 인터넷에 python3 html parcer 검색: html 파일 분석하는 내용 검색 
8. PyPI 검색해서 맞는 패키지 찾기 
9. pip install beautifulsoup4 다운 
10. r.text 로 리퀘스트한 내용을 html 분석모듈 parcer로 가져옴 
11. html 내용의 title만 골라서 인쇄 .sting은 태그없이 내용만 출력
12. html body중에서 div태그에 있는 class이름이 "h_movie2"인 모든 라인 골라오기 


'''

import requests 
r = requests.get('https://movie.naver.com/movie/bi/pi/basic.nhn?code=9679') ## 5. 다운받고 싶어하는 웹 홈페이지 주소 넣기 
print(r.text) ## 6. html text 파일로 인쇄

from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text,'html.parcer') ## 10. r.text 로 리퀘스트한 내용을 html 분석모듈 parcer로 가져옴 
print(soup.title.string) ## 11. html 내용의 title만 골라서 인쇄 .sting은 태그없이 내용만 출력

articles = soup.findAll('div', {'class':'h_movie2'}) ## 12.html body중에서 div태그에 있는 class이름이 "h_movie2"인 모든 라인 골라오기 
print(articles[0].text)
```
